```python
y = np.array(y)
df_y = pd.DataFrame({'Content': y })
df_y = df_y[[ 'Content']]
df_y
```

```python
print(df_y.columns.values)
```

```python
#change text to lowercase
df_y["Content"] = df_y["Content"].str.lower()
df_y["Content"]
```

```python
stop_words_list = stopwords.words("english")
def text_cleanning(text):
    
    # All characters in the text are in lowercase.
    text = str(text)
    text = text.lower()
    
    #All html tags and attributes (i.e., /<[^>]+>/) are removed
    regex_pattern = re.compile('[^A-Za-z0-9\s-]')
    text = re.sub(regex_pattern, ' ', text)
    
    #Remove duplicate spaces and strp front and end spaces
    text = re.sub(' +',' ',text)
    text = text.strip()
    
    #make text to a list of words
    text_list = text.split(' ')
    
    # All stop words are removed. 
    text_drop_stop_words_list = [x for x in text_list if x not in stop_words_list]
    
    text = " ".join(text_drop_stop_words_list)
    return text
```

```python
df_y['Cleaned_Content'] = df_y['Content'].apply(text_cleanning)

#Apply Lemmatization
df_y['Cleaned_Content'] = df_y['Cleaned_Content'].apply(lambda x: WordNetLemmatizer().lemmatize(x,'v'))

df_y= df_y.dropna()
df_y.head()
```

```python
print("The size of dataset after cleaning is:\n",df_y.shape)
```

```python
top_n_words = 50
```

```python
tokens = ' '.join(df_y['Cleaned_Content']).split(' ')
fd = FreqDist(tokens)
word_freq = pd.DataFrame(list(fd.items()), columns = ["Word", "Frequency"])\
.sort_values(by = 'Frequency', ascending = False)
top_50_words = word_freq[:top_n_words]
top_50_words.head()
```
